```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=0.3,min_samples=0.3,metric='euclidean')
```

```python
model1=db.fit(x1)
model2=db.fit(x2)
model3=db.fit(x3)
model4=db.fit(x4)
model5=db.fit(x5)
```

```python
label1=model1.labels_
label2=model2.labels_
label3=model3.labels_
label4=model4.labels_
label5=model5.labels_
```

```python
#identifying the points which makes up our core points
sample_cores=np.zeros_like(label1,dtype=bool)

sample_cores[db.core_sample_indices_]=True

#Calculating the number of clusters

n_clusters=len(set(label1))- (1 if -1 in label1 else 0)
print('No of label1 clusters:',n_clusters)
```
